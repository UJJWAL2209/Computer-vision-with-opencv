```python
# loading libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import os
```

```python
# LOADING THE IMAGE
image_path =  "C:\\Users\ASUS\\Downloads\\PICTURES\\pexels-nothing-ahead-8066665.jpg"
img = cv2.imread(image_path)
```

```python
# Convert from BGR to RGB (OpenCV reads images in BGR format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Resize the image using OpenCV
new_width = 1000
new_height = 600  
resized_img = cv2.resize(img, (new_width, new_height))
```

```python
# Convert from BGR to RGB (OpenCV reads images in BGR format)
resized_img_rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)

# Display the resized image using matplotlib
plt.imshow(resized_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Adjust brightness and contrast
alpha = 1.5  # Contrast control (1.0 means no change)
beta = 30    # Brightness control (0 means no change)

adjusted_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
adjusted_img_rgb = cv2.cvtColor(adjusted_img, cv2.COLOR_BGR2RGB)

# Display the adjusted image using matplotlib
plt.imshow(adjusted_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Rotate the image by 270 degrees (90 degrees clockwise)
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
rotated_img_rgb = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)

# Display the rotated image using matplotlib
plt.imshow(rotated_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Flip the image horizontally
flipped_horizontal = cv2.flip(img, 1)

# Flip the image vertically
flipped_vertical = cv2.flip(img, 0)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
flipped_horizontal_rgb = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flipped_vertical_rgb = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)

# Display the flipped images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontal_rgb)
plt.title('Flipped Horizontally')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_vertical_rgb)
plt.title('Flipped Vertically')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply a Laplacian filter for sharpening
sharpened_img = cv2.Laplacian(img, cv2.CV_8U)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
sharpened_img_rgb = cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB)

# Display the sharpened image using matplotlib
plt.imshow(sharpened_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Define the desired color (BGR format)
desired_color = (0, 255, 0)  # Green color in BGR

# Create a new image with the desired color
colored_img = np.zeros_like(img)
colored_img[:] = desired_color

# Blend the original image with the colored image
alpha = 0.5  # Blending factor
blended_img = cv2.addWeighted(img, 1 - alpha, colored_img, alpha, 0)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
blended_img_rgb = cv2.cvtColor(blended_img, cv2.COLOR_BGR2RGB)

# Display the blended image using matplotlib
plt.imshow(blended_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using matplotlib
plt.imshow(gray_img, cmap='gray')  # Use 'gray' colormap for grayscale images
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Calculate the dimensions of the middle portion
height, width = img.shape[:2]
crop_top = height // 4
crop_bottom = 3 * height // 4
crop_left = width // 4
crop_right = 3 * width // 4

# Crop the middle portion of the image
cropped_img = img[crop_top:crop_bottom, crop_left:crop_right]

# Convert from BGR to RGB (OpenCV reads images in BGR format)
cropped_img_rgb = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB)

# Display the cropped image using matplotlib
plt.imshow(cropped_img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
# Directory containing the images
image_directory = "C:\\Users\\ASUS\\Downloads\\PICTURES"

# List all image files in the directory
image_files = [file for file in os.listdir(image_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Select a random image file
random_image_file = random.choice(image_files)

# Load the randomly selected image using OpenCV
random_image_path = os.path.join(image_directory, random_image_file)
img = cv2.imread(random_image_path)

# Convert from BGR to RGB (OpenCV reads images in BGR format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the randomly selected image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Turn off axis labels and ticks
plt.title(f"Random Image: {random_image_file}")
plt.show()
```
